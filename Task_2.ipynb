```python
import numpy as np

# Task 2: Calculus - Population Growth Analysis
print("\nTask 2: Calculus - Population Growth Rate")
print("=" * 45)

# Population function: P(t) = 1000 * e^(0.03t)
P0 = 1000  # Initial population
r = 0.03   # Growth rate
t_max = 20 # Time period

# Calculate population at different times
t_values = np.array([0, 5, 10, 15, 20])
populations = P0 * np.exp(r * t_values)
growth_rates = r * populations  # dP/dt = r * P(t)

print("Time (years)\tPopulation\tGrowth Rate")
print("-" * 45)
for i, t in enumerate(t_values):
    print(f"{t:>4.0f}\t\t{populations[i]:>8.0f}\t{growth_rates[i]:>8.2f}")

# Maximum growth rate (at t = 20)
P_20 = P0 * np.exp(r * 20)
max_growth_rate = r * P_20

print(f"\nMaximum Growth Analysis:")
print(f"Population at t=20: {P_20:.1f} people")
print(f"Maximum growth rate: {max_growth_rate:.2f} people/year")
print(f"Growth factor over 20 years: {P_20/P0:.2f}x")
```
