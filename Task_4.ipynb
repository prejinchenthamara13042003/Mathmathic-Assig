```python
import numpy as np
import scipy.stats as stats
import math

# Task 4: Statistics - District Population Confidence Interval
print("\nTask 4: Statistics - 95% Confidence Interval")
print("=" * 50)

# District population data
data = np.array([1050, 1070, 1065, 1080, 1045, 1075, 1060, 1070])
n = len(data)

# Calculate descriptive statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
sample_var = np.var(data, ddof=1)

print("District Population Data:")
print(data)
print(f"\nDescriptive Statistics:")
print(f"Sample size (n):      {n}")
print(f"Sample mean (x̄):      {sample_mean:.2f}")
print(f"Sample std dev (s):   {sample_std:.4f}")
print(f"Sample variance (s²): {sample_var:.4f}")

# 95% Confidence Interval calculation
confidence_level = 0.95
alpha = 1 - confidence_level
df = n - 1

# t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

# Standard error and margin of error
standard_error = sample_std / math.sqrt(n)
margin_of_error = t_critical * standard_error

# Confidence interval bounds
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"\n95% Confidence Interval Calculation:")
print(f"Degrees of freedom:   {df}")
print(f"t-critical value:     {t_critical:.4f}")
print(f"Standard error:       {standard_error:.4f}")
print(f"Margin of error:      {margin_of_error:.2f}")
print("-" * 40)
print(f"95% CI: ({lower_bound:.2f}, {upper_bound:.2f})")
print(f"Interpretation: We are 95% confident the true mean")
print(f"district population is between {lower_bound:.0f} and {upper_bound:.0f}")
```
