```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math

# Task 1: Linear Algebra - Population Interaction Matrix
print("Task 1: Linear Algebra - Population Dynamics")
print("=" * 50)

# Given transition matrix
A = np.array([[1.2, 0.1],
              [0.05, 1.1]])

print("Transition Matrix A:")
print(A)

# Calculate eigenvalues and eigenvectors
eigenvals, eigenvecs = np.linalg.eig(A)

# Sort by eigenvalue magnitude (largest first)
idx = np.argsort(eigenvals.real)[::-1]
eigenvals = eigenvals[idx]
eigenvecs = eigenvecs[:, idx]

# Normalize eigenvectors (make second component = 1)
v1 = eigenvecs[:, 0].real / eigenvecs[:, 0].real[1]
v2 = eigenvecs[:, 1].real / eigenvecs[:, 1].real[1]

print(f"\nEigenvalues:")
print(f"λ₁ = {eigenvals[0].real:.6f}")
print(f"λ₂ = {eigenvals[1].real:.6f}")

print(f"\nEigenvectors:")
print(f"v₁ = ({v1[0]:.3f}, {v1[1]:.0f})")
print(f"v₂ = ({v2[0]:.3f}, {v2[1]:.0f})")

print(f"\nStability Analysis:")
print(f"Both eigenvalues > 1 → Both populations will GROW")
print(f"Dominant eigenvalue: λ₁ = {eigenvals[0].real:.3f}")
```
