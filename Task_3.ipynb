```python
import math

# Task 3: Probability - Immigration Events (Poisson Distribution)
print("\nTask 3: Probability - Poisson Distribution Analysis")
print("=" * 55)

# Poisson parameter
lambda_val = 5

# Calculate expected value and variance
expected_value = lambda_val
variance = lambda_val
std_deviation = math.sqrt(lambda_val)

print(f"Immigration Events Analysis (λ = {lambda_val}):")
print("-" * 40)
print(f"Expected value E[X] = {expected_value}")
print(f"Variance Var(X)     = {variance}")
print(f"Standard deviation  = {std_deviation:.4f}")

# Calculate some key probabilities
prob_exactly_5 = (math.e**(-lambda_val) * lambda_val**5) / math.factorial(5)
print(f"\nKey Probabilities:")
print(f"P(X = 5) = {prob_exactly_5:.4f}")

# Show probability distribution for k = 0 to 10
print(f"\nProbability Distribution:")
print("Events (k)\tP(X = k)")
print("-" * 25)
for k in range(0, 11):
    prob = (math.e**(-lambda_val) * lambda_val**k) / math.factorial(k)
    print(f"{k:>6}\t\t{prob:.4f}")
```
